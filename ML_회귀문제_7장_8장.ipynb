# Scikit-learn 라이브러리를 이용한 회귀분석 문제 모델링
 - LinearRegression()
    - 최소제곱법(Ordinary Least Squares, OLS)을 사용하여 수학공식을 통해 가중치와 편향을 찾아 최적의 선을 변환
    - 경사하강법을 사용하지않음

```python
from sklearn import linear_model
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_squared_error

# 1. 데이터 준비 (지도학습)
# X(특징): 키(cm), y(라벨, 정답, 타겟): 몸무게(kg)
X = [[163], [179], [166], [169], [171]] # 2차원 배열
y = [54, 63, 57, 56, 58]

# 2. (단순) 선형회귀 모델 생성
regr = linear_model.LinearRegression()

# 3. 모델 학습
regr.fit(X, y)

# 4. 학습된 모델의 성능 평가 및 파라미터 확인
score = regr.score(X, y)
coef = regr.coef_ # 기울기 파라미터 (w)
print(coef)
intercept = regr.intercept_ # 절편 파라미터 (b)
print(intercept)
print(f"회귀식: y = {coef[0]:.2f} * X + {intercept:.2f}")
print(f"데이터와 선형 회귀 직선의 관계점수(R2): {score:.1%}")
print(f"회귀선의 실제 데이터 분포의 설명력: {r2_score(y, regr.predict(X))}")
print(f"회귀선의 오차 MSE: {mean_squared_error(y, regr.predict(X))}")
# 새로운 키값으로 몸무게 예측하기
un_seen = [[167]] # 2차원 배열
result = regr.predict(un_seen)
print(f"새로운 데이터 {un_seen[0][0]}cm의 추정 몸무게는 {result[0]:.1f}kg입니다.")

# 결과 시각화
plt.scatter(X, y, color = 'blue', label = 'actual data') # 실제 데이터 분포
hat_y = regr.predict(X) # 추정된 선형 회귀 직선
plt.plot(X, hat_y, label = 'regression line', color = 'red') # 회기 선 그리기
plt.xlabel("Height(cm)")
plt.ylabel("Weight(kg)")
plt.title("Linear Regression")
plt.legend()
plt.show()
```
